```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import xarray as xr
import numpy as np
```

```python
bathymetry_data = xr.open_dataset('../data/config/bathymetry/GEBCO_2023_sub_ice_topo.nc')['elevation']
```

```python
bathymetry_data
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5),
                                subplot_kw={'projection': ccrs.PlateCarree()})
extent = [-20, 20, -60, 0]  # Längengrad, Breitengrad, von Westen, Osten, Süden und Norden
ax1.set_extent(extent, crs=ccrs.PlateCarree())
ax1.add_feature(cfeature.LAND, edgecolor='black')
ax1.coastlines(resolution='10m', color='black')
bathymetry_data.plot(ax=ax1, cmap='viridis', transform=ccrs.PlateCarree())
ax2.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())  # Ganze Weltkarte
ax2.add_feature(cfeature.LAND, edgecolor='black')
ax2.coastlines(resolution='10m', color='black')
ax2.plot(extent[:2], extent[2:], 'ro-', transform=ccrs.PlateCarree())  # Zeichnen Sie den Ausschnitt
plt.show()
```
